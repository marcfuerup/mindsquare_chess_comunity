# Mindsquare Schach Rating

Hier werden die gängigen Ratings aller Mindsquarianer eingetragen die in der .csv Datei stehen.

Am Ende werden die Spielenden auch nach Rating sortiert.

```python
import pandas as pd
import requests

df = pd.read_csv('ratings.csv')
```

Bis hier wurden nur Bibliotheken und die CSV der Spielenden geladen 
Jetzt kommt der API-Call an Chess.com. Die Daten werden in einem DataFrame geladen.

```python
def get_chess_ratings(username):
    headers = {
        "User-Agent": "Mozilla/5.0 (compatible; MyChessStatsBot/1.0)"
    }
    url = f'https://api.chess.com/pub/player/{username}/stats'
    try:
        response = requests.get(url, headers=headers, timeout=500)
        response.raise_for_status()
        data = response.json()

        blitz = data.get('chess_blitz', {}).get('last', {}).get('rating')
        rapid = data.get('chess_rapid', {}).get('last', {}).get('rating')
        bullet = data.get('chess_bullet', {}).get('last', {}).get('rating')

        return blitz, rapid, bullet

    except requests.RequestException:
        return None, None, None

df[['Blitz', 'Rapid', 'Bullet']] = df['Username'].apply(
    lambda u: pd.Series(get_chess_ratings(u))
)

df
```

Jetzt folgt nur noch eine Sortierung nach dem höchsten Rating unabhängig von Kategorie

```python
df[['Blitz', 'Rapid', 'Bullet']] = df[['Blitz', 'Rapid', 'Bullet']].apply(pd.to_numeric, errors='coerce')
df['MaxRating'] = df[['Blitz', 'Rapid', 'Bullet']].max(axis=1)

df_sorted = df.sort_values(by='MaxRating', ascending=False)
df_sorted[['Name', 'Username', 'Blitz', 'Rapid', 'Bullet', 'MaxRating']]
```
